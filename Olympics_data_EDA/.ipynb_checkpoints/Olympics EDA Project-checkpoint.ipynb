```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
athletes = pd.read_csv('athlete_events.csv')
athletes
```

```python
regions = pd.read_csv('noc_regions.csv')
regions
```

```python
athletes.shape  , regions.shape
```

### Merging the 02 data frames using a left join

```python
athletes_data = pd.merge(left = athletes , right = regions , how = 'left' , on ="NOC"  )
athletes_data.head(10)
```

```python
# The names of the columns are inconsistant. So the column names are converted to upper case
athletes_data.columns = athletes_data.columns.str.upper()
athletes_data
```

```python
athletes_data.info()
```

```python
athletes_data.describe()
```

### Checking For null values

```python
athletes_data.isnull().sum()
```

```python
athletes_data.isnull().sum()/len(athletes_data)*100
```

* The dataset contains several columns with missing and nan values
* Higher percentage of Missing values in Medal Column is acceptable.
* Missing values in columns Age , Height , weight is replaced with the mean
* Missing values from the Region column is replaced with TEAM name
* and missing values in medal column is replaced with 'No medal'



```python
athletes_data['AGE'].fillna((athletes_data['AGE'].mean()),inplace=True)
athletes_data['HEIGHT'].fillna((athletes_data['HEIGHT'].mean()),inplace=True)
athletes_data['WEIGHT'].fillna((athletes_data['WEIGHT'].mean()),inplace=True)
```

```python
athletes_data['REGION'].fillna(athletes_data['TEAM'],inplace=True)
athletes_data['MEDAL'].fillna('No medal',inplace=True)
```

```python
athletes_data.isnull().sum()
```

```python
# Visualizing the Age Distribution
plt.figure(figsize = (12 , 6))
plt.title("Age Distribution of the athletes")
plt.xlabel("Age")
plt.ylabel("Number")
plt.hist(athletes_data['AGE'] , 
         bins = np.arange(10 , 80 , 2) ,
         color = 'red' , 
         edgecolor = 'black');
```

* Most of the athletes are in the age range of 20 to 30.
* several athletes are above age of 50 and even 60
* Very younger athletes below the age of 18 also present

```python
# Gender Analysis
gender = athletes_data['SEX'].value_counts()
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender, labels=gender.index, autopct='%1.1f%%')
plt.show()
```

```python
# Different sports events in olympics
athletes_data['SPORT'].unique()
```

```python
# Analyzing the number of participants per each year of olympics
numberof_ath = athletes_data.groupby('YEAR')['ID'].nunique()
plt.figure(figsize=(15,10))
numberof_ath.plot(kind='barh' , color = 'orange')
plt.show();
```

* The Highest number of athletes were recorded in 1992 and 2016 Olympics
* Lowest number is recorded in the First ever olympics in 1896

## Analysis of medals

```python
medal_count = athletes_data['MEDAL'].value_counts()
medal_count
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = medal_count.index , y = medal_count , palette = 'Set3' )
plt.show()
```

```python
# Number of medals per year
medal_per_year = athletes_data.groupby(['YEAR'])['MEDAL'].value_counts().reset_index(name='medal_per_year')
medal_count_year =pd.pivot_table(medal_per_year, values='medal_per_year', index=['YEAR'], columns=['MEDAL'], aggfunc=np.sum)[:10]
medal_count_year
```

## Distribution of medals with gender

```python
# Gender wise meda distribution for all olympics
medal_dist = athletes_data[athletes_data['MEDAL'] != 'No medal']
plt.figure(figsize=(6,4))
sns.countplot(x='MEDAL',hue='SEX', data=medal_dist)
plt.xlabel('Medal')
plt.ylabel('Number of Medals')
plt.title('Gender wise medal distribution in Olympics')
plt.show();
```

### Gender wise medal distribution for Summer olympics

```python
summer_data=athletes_data[athletes_data['SEASON']=='Summer']
sum_meds = summer_data[summer_data['MEDAL'] != 'No medal']
plt.figure(figsize=(6,4))
sns.countplot(x='MEDAL',hue='SEX', data=sum_meds)
plt.xlabel('Medal')
plt.ylabel('Number of Medals')
plt.title('Medal Distribution as per gender in Summer Olympics')
plt.show();
```

### Gender wise meda distribution for Winter olympics

```python
winter_data=athletes_data[athletes_data['SEASON']=='Winter']
wint_meds = winter_data[winter_data['MEDAL'] != 'No medal']
plt.figure(figsize=(6,4))
sns.countplot(x='MEDAL',hue='SEX', data=wint_meds)
plt.xlabel('Medal')
plt.ylabel('Number of Medals')
plt.title('Medal Distribution as per gender in Winter Olympics')
plt.show();
```

### Analysis of the number of  participants based on gender per year

```python
num_men = athletes_data[athletes_data['SEX'] == 'M'].groupby('YEAR').count()['NAME'].reset_index()
num_women = athletes_data[athletes_data['SEX'] == 'F'].groupby('YEAR').count()['NAME'].reset_index()

men_women_Total = num_men.merge(num_women ,on = 'YEAR' , how = 'left')
men_women_Total.rename(columns={'NAME_x' : 'Male' , 'NAME_y' : 'Female'} , inplace = True)
men_women_Total = men_women_Total.fillna(0)
men_women_Total.head()
```

### participants of Women per year in Summer olympics

```python
women_counts = athletes_data[(athletes_data['SEX'] == 'F') & (athletes_data['SEASON'] == 'Summer')]
sns.set(style = "darkgrid")
plt.figure(figsize=(15,8))
sns.countplot(x = 'YEAR' , data = women_counts)
sns.color_palette("pastel")
plt.title('Woman Participation per year in Summer olympics')
plt.show();
```

```python
women_trend = women_counts.groupby('YEAR')['SEX'].value_counts()
plt.figure(figsize =(15,8) )
women_trend.loc[:,'F'].plot()
plt.title("Plot of Female Athletes in Summer Olympics over time")
plt.show();
```

### participants of men per year in Summer olympics

```python
men_counts = athletes_data[(athletes_data['SEX'] == 'M') & (athletes_data['SEASON'] == 'Summer')]
sns.set(style = "darkgrid")
plt.figure(figsize=(15,8))
sns.countplot(x = 'YEAR' , data = men_counts , palette = "Spectral" )
plt.title('Male Participation per year in Summer Olympics')
plt.show();
```

```python
male_trend = men_counts.groupby('YEAR')['SEX'].value_counts()
plt.figure(figsize =(15,8))
male_trend.loc[:,'M'].plot()
plt.title("Plot of Male Athletes over time")
plt.show();
```
